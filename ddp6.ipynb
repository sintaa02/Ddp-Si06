```python
#SOAL 2 Buat lah output dari menggunakan bahasa pemprograman python dengan :
jumlah= 0 #untuk menampung hasil jumlah
string = "" #untuk menampung hasil jumlah dalam bentuk string
bilangan= 1
while bilangan <=19: #melakukan perulangan sebanyak <= 19
   jumlah += bilangan
   string += str(bilangan)
   if bilangan < 19 :
      string += "+"
   bilangan += 2
print (f"{string} = {jumlah}")
```
